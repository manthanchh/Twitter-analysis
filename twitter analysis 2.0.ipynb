FOR EACH DOCUMENT IN DOCUMENT ARRAY, REMOVE STOPWORDS AND PERFROM LEMMATIZATION ON EACH WORD

Lemmatization

```python
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import sentiwordnet as swn, wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

Remove : Stopwords + punctuations

```python
stopwrd = stopwords.words('english')
import string
punctuations = list(string.punctuation)
stopwrd =  stopwrd + punctuations 
```

```python
def clean_tweets(words):
    output_tweet = []
    for t in words:
        if t.lower() not in stopwrd:
            pos = pos_tag([t])
            clean_tweet = lemmatizer.lemmatize(t,pos = get_simple_pos(pos[0][1]))
            output_tweet.append(clean_tweet.lower())
    return output_tweet
```

```python
tweets = [(clean_tweets(text),category) for text,category in tweets ]
```

Build feature set:

1. We will first create an array which contain all the words present in all the text docs.
2. Find Frequencyof each word
3. Take the top frequent n words as a feature.
4. Then se whether those top frequnet words present in the document or not.

```python
all_words = []
for doc in tweets:
    all_words = all_words + doc[0]
all_words
```

Find the frequency distribution:

```python
import nltk
Freq_Dist = nltk.FreqDist(all_words)
common_Feature = Freq_Dist.most_common(2500)
```

Our Important Features are:

```python
features = [i[0] for i in common_Feature]
features
```

Now we have our features. We will create a dictionary of whether the feature word is present in the particular tweet text or not. It it is present, it return true or else false. This dictionary will be present in the particular tuple along with its category(output).

Create a function to get the feature dictionary:

```python
def feature_dict(words):
    current_feature = {}
    words_set = set(words)
    for w in features:
        current_feature[w] = w in words_set
    return current_feature
```
